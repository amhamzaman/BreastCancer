```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import time
from subprocess import check_output
print(check_output(["dir","data"], shell = True))

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("../BreastCancer/data/data.csv")
```

```python
data.head()
```

```python
y = data.diagnosis
x = data.drop(["id", "Unnamed: 32", "diagnosis"], axis = 1)
x.head()
```

```python
plt.hist(y, 3, color = ['#79cdcd'])
B, M = y.value_counts()
print('Number of Benign: ', B)
print('Number of Malignant: ', M)
```

```python
x.describe()
```

```python
dat = x
dat = (dat - dat.mean()) / dat.std()
dat = pd.concat([y, dat.iloc[:,0:10]], axis = 1)
dat = pd.melt(dat, id_vars= "diagnosis", var_name= "features", value_name= "value")

plt.figure(figsize = (15,10))
sns.violinplot(x="features", y= "value", hue="diagnosis", data= dat, split= True, inner="quart")
plt.xticks(rotation= 45)

plt.figure(figsize=(10,10))
sns.boxplot(x= "features", y= "value", hue= "diagnosis", data= dat)
plt.xticks(rotation= 45)
```

```python
sns.jointplot(x='concavity_worst', y='concave points_worst', data=x, kind='reg')
```
